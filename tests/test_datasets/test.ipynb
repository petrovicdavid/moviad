```python
from moviad.datasets.mvtec import MVTecDataset
from moviad.datasets.dataset_arguments import DatasetArguments
import numpy as np
import torch

args = DatasetArguments(
    dataset_path="/mnt/mydisk/manuel_barusco/datasets/mvtec",
    category="bottle",
    split="test",
    img_size=(256, 256),
    gt_mask_size=(256, 256),
    image_transform_list=None,
)

dataset = MVTecDataset(args)
dataset.load_dataset()
```

```python
dataset[0][1]
```

```python
dataloader = torch.utils.data.DataLoader(dataset, batch_size=4, shuffle=False)  
```

```python
from tqdm import tqdm
for batch in tqdm(dataloader):
    images, labels, gt_masks, paths = batch
    print(f"Images shape: {images.shape}")
    print(f"GT Masks shape: {gt_masks.shape}")
    break
```
